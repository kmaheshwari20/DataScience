```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv ("/Users/kapil.maheshwari/Downloads/National_Stock_Exchange_of_India_Ltd.csv", delimiter = ",")
```

```python
df.info()
```

```python
df[df['Symbol'] == 'ADANIPORTS']
```

```python
df['Open'] = pd.to_numeric(df['Open'].str.replace(',',''))
df['High'] = pd.to_numeric(df['High'].str.replace(',',''))
df['Low'] = pd.to_numeric(df['Low'].str.replace(',',''))
df['LTP'] = pd.to_numeric(df['LTP'].str.replace(',',''))
df['Turnover (crs.)'] = pd.to_numeric(df['Turnover (crs.)'].str.replace(',',''))
df['52w H'] = pd.to_numeric(df['52w H'].str.replace(',',''))
df['52w L'] = pd.to_numeric(df['52w L'].str.replace(',',''))
```

```python
df.info()
```

```python
df.head()
```

```python
df.rename(columns={'Open':'last_open_price', 'High': 'last_high_price','Low':'last_low_price','LTP':'last_trading_price',
                   'Chng':'Change','% Chng': 'percentage_change','Volume (lacs)':'volume_in_lacs', 
                   'Turnover (crs.)':'turnover_in_crs', '52w H':'52w_high','52w L':'52w_low',
                   '365 d % chng':'365d_perc_chng','30 d % chng':'30d_perc_chng'
                  }, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='last_open_price', data=df, palette='plasma', order=df.sort_values('last_open_price', ascending = False).Symbol)
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='percentage_change', data=df, palette='plasma', order=df.sort_values('percentage_change', ascending = False).Symbol)
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='turnover_in_crs', data=df, palette='plasma', order=df.sort_values('turnover_in_crs', ascending = False).Symbol)
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='52w_high', data=df, palette='plasma', order=df.sort_values('52w_high', ascending = False).Symbol)
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='365d_perc_chng', data=df, palette='plasma', order=df.sort_values('365d_perc_chng', ascending = False).Symbol)
```

```python
plt.figure(figsize=(20, 15))
plt.xticks(rotation=90)
sns.barplot(x='Symbol', y='30d_perc_chng', data=df, palette='plasma', order=df.sort_values('30d_perc_chng', ascending = False).Symbol)
```

```python
corr_matrix = df.corr()
corr_matrix['30d_perc_chng'].sort_values(ascending=False)
```

```python
#heatmap
matrix = np.triu(corr_matrix)
plt.figure(figsize=(15, 11))
sns.heatmap(corr_matrix, annot=True, cmap = 'RdPu')
```
